### 9. Simulating Diffusion Denoising

```python
# Import numpy for numerical operations and matplotlib for plotting images
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 28x28 noisy image with random values between 0 and 1
image = np.random.rand(28, 28)
# Display the noisy image in grayscale
plt.imshow(image, cmap='gray')
plt.title("Step 0: Noise")
plt.show()
```

```python
# Simulate denoising by reducing noise in each step
for step in range(1, 4):
    image = image * 0.9  # reduce noise by scaling pixel values
    plt.imshow(image, cmap='gray')
    plt.title(f"Step {step}: Denoising")
    plt.show()
```
